## Create a ndarray of images with ImageCollection

```python
from skimage import io
```

```python
image_array = io.ImageCollection('test*.jpg')
```

```python
len(image_array)
```

```python
type(image_array)
```

```python
image_idarray = io.concatenate_images(image_array)
```

```python
type(image_idarray)
```

```python
from skimage.feature import corner_fast, corner_peaks
from skimage.color import rgb2gray
```

```python
grag_image = rgb2gray(image_idarray[0])
```

```python
corner_peaks(corner_fast(grag_image), min_distance=1)
```
