```python
config = {
    'host': 'localhost',
    'user': 'root',
    'password': '123456',
    'db': 'prod_schema',
    'charset': 'utf8mb4',
    'cursorclass': pymysql.cursors.DictCursor
}
def db_execute(config, sql):
    connection = pymysql.connect(**config)
    try:
        with connection.cursor() as cursor:
            cursor.execute(sql)
        connection.commit()
        return True
    finally:
        connection.close()
```

```python
def update_cluster(config, row_data):
    img_path = row_data['img_path']
    cid_5 = row_data['cid_5']
    cid_15 = row_data['cid_15']
    cid_25 = row_data['cid_25']
    sql  = "update prod_schema.prod_img set cid_5 = %d, cid_15 = %d, cid_25 = %d \
where img_path = '%s'" % (cid_5, cid_15, cid_25, img_path)
    #print sql
    db_execute(config, sql)
```

```python
update_cluster(config, feature_sf[0])
```

```python
for i in range(len(feature_sf)):
    row_data = feature_sf[i]
    update_cluster(config, row_data)
```
